### Filtering a DataFrame

```python
ipl.head()
```

```python
# find all the final winners
```

```python
ipl[ipl["MatchNumber"] == "Final"]
```

```python
ipl[ipl["MatchNumber"] == "Final"][["Season","WinningTeam"]]
```

```python
# how many superover finishes have occured
```

```python
ipl[ipl["SuperOver"] == "Y"]
```

```python
len(ipl[ipl["SuperOver"] == "Y"])
```

```python
# how many matches have csk won in kolkata
```

```python
ipl[(ipl["City"] == "Kolkata") & (ipl["WinningTeam"] == "Chennai Super Kings")].head()
```

```python
len(ipl[(ipl["City"] == "Kolkata") & (ipl["WinningTeam"] == "Chennai Super Kings")])
```

```python
# toss winner is match winner in percentage
```

```python
len(ipl[ipl["TossWinner"] == ipl["WinningTeam"]])/ipl.shape[0]
```

```python
# movies with rating higher than 8 and votes > 10000
```

```python
movies.head()
```

```python
movies[(movies["imdb_rating"] > 8) & (movies["imdb_votes"]>10000)].head()
```

```python
len(movies[(movies["imdb_rating"] > 8) & (movies["imdb_votes"]>10000)])
```

```python
# Action movies with rating higher than 7.5
```

```python
movies[(movies["genres"].str.contains("Action",case=False)) & (movies["imdb_rating"] > 7.5)].head()
```

```python
len(movies[(movies["genres"].str.contains("Action",case=False)) & (movies["imdb_rating"] > 7.5)])
```
