### Series is 1D and DataFrames are 2D objects

- But why?
- And what exactly is index?

```python
# can we have multiple index? Let's try
index_val = [('cse',2019),('cse',2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]
a = pd.Series([1,2,3,4,5,6,7,8],index=index_val)
a
```

```python
a[("cse",2022)]
```

```python
# The problem?
a["cse"]
```

```python
# The solution -> multiindex series(also known as Hierarchical Indexing)
# multiple index levels within a single index
```

```python
# how to create multiindex object
# 1. pd.MultiIndex.from_tuples()
# 2. pd.MultiIndex.from_product()
```

```python
# 1. pd.MultiIndex.from_tuples()
```

```python
index_val = [('cse',2019),('cse',2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]
multiindex = pd.MultiIndex.from_tuples(index_val)
multiindex
```

```python
type(multiindex)
```

```python
multiindex.levels
```

```python
# level inside multiindex object
```

```python
multiindex.levels[0]
```

```python
multiindex.levels[1]
```

```python
# 2. pd.MultiIndex.from_product()
```

```python
pd.MultiIndex.from_product([["cse","ece"],[2019,2020,2021,2022]])
```

```python
# creating a series with multiindex object
```

```python
s = pd.Series([1,2,3,4,5,6,7,8],index=multiindex)
s
```

```python
# how to fetch items from such a series
```

```python
s[("cse",2022)]
```

```python
s["cse"]
```

```python
# unstack 
# multiindex to dataframe
```

```python
s
```

```python
temp = s.unstack()
temp
```

```python
# stack
# dataframe to multiindex
```

```python
temp.stack()
```

### Long Vs Wide Data
image.png

**Wide format** is where we have a single row for every data point with multiple columns to hold the values of various attributes.

**Long format** is where, for each data point we have as many rows as the number of attributes and each row contains the value of a particular attribute for a given data point.

```python
# melt -> simple example branch
```

```python
# wide to long
```

```python
pd.DataFrame({"cse":[120]})
```

```python
pd.DataFrame({"cse":[120]}).melt()
```

```python
# melt -> branch with year
```

```python
pd.DataFrame({"cse":[120],"ece":[100],"mech":[50]})
```

```python
pd.DataFrame({"cse":[120],"ece":[100],"mech":[50]}).melt()
```

```python
pd.DataFrame({"cse":[120],"ece":[100],"mech":[50]}).melt(var_name="branch",value_name="num_students")
```

```python
pd.DataFrame(
{
    "branch":["cse","ece","mech"],
    "2020":[100,150,60],
    "2021":[120,130,80],
    "2022":[1550,140,70]
})
```

```python
pd.DataFrame(
{
    "branch":["cse","ece","mech"],
    "2020":[100,150,60],
    "2021":[120,130,80],
    "2022":[1550,140,70]
}).melt()
```

```python
pd.DataFrame(
{
    "branch":["cse","ece","mech"],
    "2020":[100,150,60],
    "2021":[120,130,80],
    "2022":[1550,140,70]
}).melt(id_vars=["branch"])
```

```python
pd.DataFrame(
{
    "branch":["cse","ece","mech"],
    "2020":[100,150,60],
    "2021":[120,130,80],
    "2022":[1550,140,70]
}).melt(id_vars=["branch"],var_name="year",value_name="students")
```

```python
# melt -> real world example
```

```python
confirmed = pd.read_csv("time_series_covid19_confirmed_global.csv")
death = pd.read_csv("time_series_covid19_deaths_global.csv")
```

```python
death.head()
```

```python
death.shape
```

```python
death = death.melt(id_vars=["Province/State","Country/Region","Lat","Long"],var_name="Date",value_name="Death")
death
```

```python
confirmed.head()
```

```python
confirmed = confirmed.melt(id_vars=["Province/State","Country/Region","Lat","Long"],var_name="Date",value_name="Confirmed")
confirmed
```

```python
confirmed.merge(death,on=["Province/State","Country/Region","Lat","Long","Date"])
```

```python
confirmed.merge(death,on=["Province/State","Country/Region","Lat","Long","Date"])[["Country/Region","Date","Confirmed","Death"]]
```

### Pivot Table

The pivot table takes simple column-wise data as input, and groups the entries into a two-dimensional table that provides a multidimensional summarization of the data.

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.groupby("sex")["total_bill"].mean()
```

```python
df.groupby(["sex","smoker"])["total_bill"].mean()
```

```python
df.groupby(["sex","smoker"])["total_bill"].mean().unstack()
```

```python
df.pivot_table(index="sex",columns="smoker",values="total_bill")
```

```python
# aggfunc
```

```python
df.pivot_table(index="sex",columns="smoker",values="total_bill",aggfunc="sum")
```

```python
df.pivot_table(index="sex",columns="smoker",values="total_bill",aggfunc="count")
```

```python
df.pivot_table(index="sex",columns="smoker",values="total_bill",aggfunc="std")
```

```python
# all cols together
```

```python
df.pivot_table(index="sex",columns="smoker")
```

```python
df.pivot_table(index="sex",columns="smoker")["tip"]
```

```python
# multidimensional
```

```python
df
```

```python
df.pivot_table(index=["sex","smoker"],columns=["day","time"],values="total_bill")
```

```python
df.pivot_table(index=["sex","smoker"],columns=["day","time"])
```

```python
df.pivot_table(index=["sex","smoker"],columns=["day","time"],aggfunc={"size":"mean","tip":"max","total_bill":"sum"})
```

```python
# margins
```

```python
df.pivot_table(index="sex",columns="smoker",values="total_bill",aggfunc="sum")
```

```python
df.pivot_table(index="sex",columns="smoker",values="total_bill",aggfunc="sum",margins=True)
```
